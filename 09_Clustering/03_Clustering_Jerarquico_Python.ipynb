# Distance Matrix

```python
# Import Libraries
from scipy.spatial import distance_matrix
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
```

```python
# Import Dataset
data = pd.read_csv('../datasets/movies/movies.csv',sep=';')
```

```python
data
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

___No hace falta normalizar porque todas las pelÃ­culas vienen puntuadas del 0 al 10___

```python
dm1 = distance_matrix(data[movies],data[movies],p=1) # Manhattan Distance
dm2 = distance_matrix(data[movies],data[movies],p=2) # Euclidean Distance
dm10 = distance_matrix(data[movies],data[movies],p=10) # p>>> distance<<<
```

```python
# How does it look like?
dm1
```

```python
# To Dataframe
def distance_matrix_to_df(dd,col_name):
    return pd.DataFrame(dd,index=col_name,columns=col_name)
```

```python
distance_matrix_to_df(dm1,data['user_id'])
```

```python
distance_matrix_to_df(dm2,data['user_id'])
```

```python
distance_matrix_to_df(dm10,data['user_id'])
```

### 2. Clustering Jerarquico

#### 2.1 Pasos
- Encontramos la menor matriz de dsitancias y unimos las dos observaciones en un nuevo culster
- Recalculamos la matriz de distancias entre los clusters antiguos y los clusters nuevos
- Repetimos el paso 1 y 2 hasta que solo quede un cluster con N observaciones

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies],method='ward',metric='euclidean')
```

```python
Z
```

```python
fig = plt.figure(figsize=(25,10))
plt.title('Dendograma para el Clustering Jerarquico')
plt.xlabel('ID usuarios Netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=0, leaf_font_size=10)
plt.show()
fig.savefig('../../MyCheatSheets/img/ward.png');
```

```python
Z = linkage(data[movies],method='centroid',metric='euclidean')
fig = plt.figure(figsize=(25,10))
plt.title('Dendograma para el Clustering Jerarquico')
plt.xlabel('ID usuarios Netflix')
plt.ylabel('Distancia')
dendrogram(Z, leaf_rotation=0, leaf_font_size=10)
plt.show()
fig.savefig('../../MyCheatSheets/img/centroid.png');
```
