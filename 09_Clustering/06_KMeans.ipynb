# KMeans

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(1234)
data = np.random.random(90).reshape(30,3)
data
```

```python
data[:,0]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(xs=data[:,0],ys=data[:,1],zs=data[:,2])
```

```python
# Definimos previamente el número de clusters que vamos a realizar k = 2 (Marcará el número de centroides)
# Elegimos Aleatoriamente Los centroides
c1 = np.random.choice(range(len(data)))
c2 = np.random.choice(range(len(data)))
clust_centers = np.vstack([data[c1],data[c2]])
```

```python
clust_centers
```

```python
from scipy.cluster.vq import vq
labels,distances = vq(data,clust_centers)
print(labels)
print(distances)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(xs=data[:,0],ys=data[:,1],zs=data[:,2],c = labels, cmap="prism")
ax.scatter(xs=clust_centers[0,0],ys=clust_centers[0,1],zs=clust_centers[0,2],c='black')
ax.scatter(xs=clust_centers[1,0],ys=clust_centers[1,1],zs=clust_centers[1,2],c='black');
```

```python
from scipy.cluster.vq import kmeans
# Return the new centroids
new_centroids = kmeans(data,clust_centers)
new_centroids
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(xs=data[:,0],ys=data[:,1],zs=data[:,2],c = labels, cmap="prism")
ax.scatter(xs=new_centroids[0][0,0],ys=new_centroids[0][0,1],zs=new_centroids[0][0,2],c='black')
ax.scatter(xs=new_centroids[0][1,0],ys=new_centroids[0][1,1],zs=new_centroids[0][1,2],c='black');
```
